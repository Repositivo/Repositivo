**To-Do:**

**Hint**

Puede ser útil los métodos ```.drop_duplicates()``` y ```.dropna()``` como también utilizar queries con máscaras booleanas.


**Respuesta Esperada:**

````python
def clean_data(dataframe):
    # Transformar a entero las notas
    dataframe["math score"] = pd.to_numeric(dataframe["math score"], errors="coerce")
    dataframe["reading score"] = pd.to_numeric(dataframe["reading score"], errors="coerce")
    dataframe["writing score"] = pd.to_numeric(dataframe["writing score"], errors="coerce")

    # Eliminar notas que no esten en el rango de 0 a 100
    dataframe = dataframe.loc[(dataframe["math score"] >= 0) & (dataframe["math score"] <= 100)]
    dataframe = dataframe.loc[(dataframe["reading score"] >= 0) & (dataframe["reading score"] <= 100)]
    dataframe = dataframe.loc[(dataframe["writing score"] >= 0) & (dataframe["writing score"] <= 100)]
    return dataframe

    # Eliminar nombres duplicados, ya que no pueden dar los mismos exámenes dos veces
    dataframe = dataframe.drop_duplicates(subset=["names"])

    # Eliminar filas duplicadas en general
    dataframe = dataframe.drop_duplicates()

    # Eliminar nulos
    dataframe = dataframe.dropna()

    return dataframe

df_grades = clean_data(df_grades)
df_grades
````

En una de las reuniones con Don Caguayo este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
  dataframe_in[columna] = (((dataframe_in[columna])*0.06) +1)
  dataframe_in[columna] = dataframe_in[columna].round(2)
  return dataframe_in
````

### 4.2 Análisis por Nivel Educacional Etnia de los Padres [0.8 Puntos]

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión:

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.

**Análisis [0.3 Puntos]**

Finalmente, responda las siguientes preguntas sobre el gráfico de caja que viene implementado a continuación. En este se muestra el GPA con respecto al nivel educacional, donde la variable de color es la etnicidad.

````python
df_grades_nueva["GPA"] = df_grades_nueva[["math score","reading score","writing score"]].mean(axis=1).round(2)


education_mapping = {
    'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    "bachelor's degree": 'college',
    "associate's degree": 'college',
    "master's degree": 'postgraduate'
}
df_grades_nueva['simple parental level of education'] = df_grades_nueva['parental level of education'].map(education_mapping)

# Paso 3: Agregar según 2 niveles: race/ethnicity y simple parental level of education
summary_table = df_grades_nueva.groupby(['race/ethnicity', 'simple parental level of education']).agg(
    GPA=('GPA', 'mean'),
    count=('GPA', 'count')
).reset_index()

total_counts = df_grades_nueva.groupby('race/ethnicity')['GPA'].count().reset_index()
total_counts.columns = ['race/ethnicity', 'count']
summary_table = pd.concat([summary_table, total_counts])

total_students = df_grades_nueva['GPA'].count()
summary_table['porcentaje'] = (summary_table['count'] / total_students * 100).map("{:.2f}%".format)

display(summary_table)
````

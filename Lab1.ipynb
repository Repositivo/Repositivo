### 1.
 Es un software de sistema de control de versiones distribuido y por ende es un sistema que registra los cambios realizado a un archivo o conjunto de los mismos. Siendo distribuido adicionalmente tiene la propiedad de estar conectado a un servidor, por lo que este registro y respaldo se mantiene de manera remota y local.
### 2. 
Git es un sistema de comandos, mientras que github entrega una interfaz gráfica. Por otro lado, git es el software que proporciona el código que estructura el sistema de control de versiones para github, el cual entrega el sistema de alojamiento web. Por último, Git proporciona también un repositorio local para los archivos y en cambio Github, es un espacio para subir la copia del repositorio de Git y además añadiendo otras funcionalidades.
### 3. 
Un hash es un una función que transforma un conjunto de elementos en un cadena de de carateres o números finitos. En lo que compete a Git los commits son snapshots, es decir una instantenea o foto, de la situación actual del conjunto de archivos, así Git crea una historial de cambios de este conjunto de todos los commits realizados, los cuales deben ser unicos e inalterables (dentro de lo posible). Para esto se utiliza un hash que apunta a los commits, para obtener un sistema key-value, es decir una llave que indique este cambio realizado que es único e inalterable (commits) y poder diferenciarlo de otras modificaciones. El hash es una cadena de 40 caracteres, y para referenciar un commit solo se necesitan unos pocos commits. Un ejemplo puede verse asi : 
 Commit             |  Author : Max
 Hash : (4015b5...) |  Parent : ###
### 4. 
Los componentes de este sistema de datos son los siguientes : 
Blob: Almacena el contenido de un objeto, archivo o cosa. Cuando un archivo es modificado, un nuevo blob almacena el archivo con las modificaciones. Un hash apunta hacia el.
Tree: Representa directorios para blobs y más árboles. Siempre hay un arbol en la raíz, apuntando a un arbol que contiene cosas. Un hash apunta hacia el.
Commits: Representa los cambios realizados a un archivo, es decir un registro, que se guarda en el historial de cambios. Un hash apunta hacia el.
### 5. 
Para inicializar un repositorio desde cero, usamos:

>> git init nombre-del-repo

Para clonar un repositorio (es decir, copiar un repositorio a tu almacenamiento local desde algún servidor github, gitlab, bitbucket u otro servidor de git):

>> git clone /ruta/al/repo.git

Para inicializar un repositorio en algún servicio de repositorios remotos, como github.

>> git status

Este permite tener una idea del estado actual del repositorio.

Por último, investigue y comente para qué sirven los siguientes comandos:

11. `git stash` (0.25 puntos)

Se utiliza para almacenar temporalmente los cambios en un directorio limpio. Una situación de uso puede, cuando se esta trabajando en una rama y se quiere cambiar de rama para realizar cambios urgentes, sin embargo, se desea guardar el estado de la rama de trabajo sin necesitar un commit, aquí es cuando se utliza git stash para almacenar los cambios locales de la rama actual. 


12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)

El comando git reset permite dehacer los cambios localmente en tu repositorio local, alterando el historial de cambios. Mientras que, git revert es utilizado para revertir los cambios especificos de un commit sin eliminar el historial de cambios, si no que, creando un nuevo commit en su lugar. Siendo útil si ya se ha compartido el trabajo con otros y no se quieren reescibir los cambios del repositorio.
